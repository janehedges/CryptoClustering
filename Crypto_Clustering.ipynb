```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
numeric_columns = [
    "price_change_percentage_24h",
    "price_change_percentage_7d",
    "price_change_percentage_14d",
    "price_change_percentage_30d",
    "price_change_percentage_60d",
    "price_change_percentage_200d",
    "price_change_percentage_1y"
]
# Apply StandardScaler to the selected columns
scaled_data = StandardScaler().fit_transform(df_market_data[numeric_columns])
```

```python
# Create a DataFrame with the scaled data
df_scaled = pd.DataFrame(scaled_data, columns=numeric_columns)

# Copy the crypto names from the original DataFrame
df_scaled["coin_id"] = df_market_data.index

# Set the coin_id column as the index
df_scaled.set_index("coin_id", inplace=True)

# Display the scaled DataFrame
df_scaled.head()
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

print(inertia)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    "k": k_values,
    "inertia": inertia
}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

df_elbow.head()
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
elbow_curve = df_elbow.hvplot.line(
    x="k",
    y="inertia",
    title="Elbow Curve to Determine Optimal k",
    xlabel="Number of Clusters (k)",
    ylabel="Inertia",
    line_width=2,
    width=800,
    height=400
)

# Display the Elbow Curve
elbow_curve
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 4

---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
best_k = 4 
kmeans_model = KMeans(n_clusters=best_k, random_state=0)
```

```python
# Fit the K-Means model using the scaled DataFrame
kmeans_model.fit(df_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame

cluster_predictions = kmeans_model.predict(df_scaled)
# Print the resulting array of cluster values.
print(cluster_predictions)
```

```python
# Create a copy of the scaled DataFrame
df_scaled_copy = df_scaled.copy()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
df_scaled_copy["cluster"] = cluster_predictions

# Display the copy of the scaled DataFrame
df_scaled_copy.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
scatter_plot = df_scaled_copy.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    c="cluster",  
    hover_cols=["coin_id"], 
    title="Cryptocurrency Clusters Based on Price Changes",
    xlabel="24h Price Change (%)",
    ylabel="7d Price Change (%)",
    width=800,
    height=400
)

# Display the scatter plot
scatter_plot
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
# down to three principal components.
pca_data = pca.fit_transform(df_scaled_copy.drop(columns=["cluster"]))

df_pca = pd.DataFrame(
    pca_data,
    columns=["PCA1", "PCA2", "PCA3"],
    index=df_scaled_copy.index
)

# View the scaled PCA data
df_pca.head()
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.
explained_variance = pca.explained_variance_ratio_
print(explained_variance)

total_explained_variance = explained_variance.sum()
print(total_explained_variance)
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 0.895 or 89.5%

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
k_values
```

```python
# Create an empty list to store the inertia values
inertia = []


# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(df_pca)
    inertia.append(kmeans.inertia_)

# Display the inertia values to verify
print(inertia)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    "k": k_values,      
    "inertia": inertia  
}
# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

df_elbow.head()
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
elbow_curve = df_elbow.hvplot.line(
    x="k",
    y="inertia",
    title="Elbow Curve to Determine Optimal k",
    xlabel="Number of Clusters (k)",
    ylabel="Inertia",
    line_width=2,
    width=800,
    height=400
)

# Display the Elbow Curve
elbow_curve
```

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Initialize the K-Means model using the best value for k
k_model = KMeans(n_clusters=4, random_state=0)
```

```python
# Fit the K-Means model using the PCA data
k_model.fit(df_pca)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled PCA DataFrame
pca_clusters = k_model.predict(df_pca)

# Print the resulting array of cluster values.
pca_clusters
```

```python
# Create a copy of the scaled PCA DataFrame
df_pca_copy = df_pca.copy()

# Add a new column to the copy of the PCA DataFrame with the predicted clusters
df_pca_copy["cluster"] = pca_clusters
# Display the copy of the scaled PCA DataFrame
df_pca_copy.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="PC1"` and `y="PC2"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
scatter_plot = df_pca_copy.hvplot.scatter(
    x="PCA1",  
    y="PCA2",  
    c="cluster",  
    hover_cols=["coin_id"], 
    title="Cryptocurrency Clusters in PCA Space",
    xlabel="PCA1",
    ylabel="PCA2",
    width=800,
    height=400
)

# Display the scatter plot
scatter_plot
```

### Visualize and Compare the Results

In this section, you will visually analyze the cluster analysis results by contrasting the outcome with and without using the optimization techniques.

```python
# Composite plot to contrast the Elbow curves
inertia_original = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(df_scaled)  
    inertia_original.append(kmeans.inertia_)

# 2. Compute the Elbow Curve With PCA 
inertia_pca = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(df_pca)  
    inertia_pca.append(kmeans.inertia_)

# 3. Create DataFrames for both curves to plot with hvplot
df_original = pd.DataFrame({"k": k_values, "inertia": inertia_original})
df_pca = pd.DataFrame({"k": k_values, "inertia": inertia_pca})

# 4. Plot the Elbow Curves using hvplot
plot_original = df_original.hvplot.line(
    x="k",
    y="inertia",
    title="Elbow Curve: Original Scaled Data vs PCA-Transformed Data",
    xlabel="Number of Clusters (k)",
    ylabel="Inertia",
    line_width=2,
    label="Original Scaled Data",
    width=800,
    height=400
)

plot_pca = df_pca.hvplot.line(
    x="k",
    y="inertia",
    xlabel="Number of Clusters (k)",
    ylabel="Inertia",
    line_width=2,
    color="red",
    label="PCA-Transformed Data",
    width=800,
    height=400
)

# Scatter to represent markers
scatter_original = df_original.hvplot.scatter(
    x="k",
    y="inertia",
    color="blue",
    size=10,
    label="Original Scaled Data",
)

scatter_pca = df_pca.hvplot.scatter(
    x="k",
    y="inertia",
    color="red",
    size=10,
    label="PCA-Transformed Data",
)

# Combine both plots
composite_plot = plot_original * plot_pca * scatter_original * scatter_pca

# Display the composite plot
composite_plot
```
